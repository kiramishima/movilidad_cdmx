```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
plt.show()
```

```python
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()
t = np.array([30, 35, 40, 50])
x = np.array([10,11,12,9])
y = np.array([50,60,70,80])

ax.plot3D(x, y, t)
ax.set_title('Costos al IRC')

# Set axes label
ax.set_xlabel('d', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('t', labelpad=20)
plt.show()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
ax.grid()
t = np.random.random(100)
x = np.random.random(100)
y = (1028/1315) * x + (9584/6575)*t + 108/263

ax.plot3D(x, y, t)
ax.set_title('Costos al IRC')

# Set axes label
ax.set_xlabel('d', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('t', labelpad=20)
plt.show()
```
